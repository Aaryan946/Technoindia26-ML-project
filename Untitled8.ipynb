```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
flower_path="/content/drive/MyDrive/CNN_Flower_Project/dataset/flowers"
images=[]
for file_name in os.listdir(flower_path):
  file_path=os.path.join(flower_path,file_name)
  img=cv2.imread(file_path)
  if img is not None:
    img =cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    img=cv2.resize(img,(224,224))
    images.append(img)
    print("images loaded:",len(images))
```

```python
images=np.array(images)
print("Shape of datasets:",images.shape)
```

```python
plt.figure(figsize=(4,4))
plt.imshow(images[0])
plt.title("Sample flower images")
plt.axis("off")
plt.show()
```

```python
images=images/255.0
print("Normalize image value from 0 to 1")
```

```python
sample_image=images[0]
print("sample image shape:",sample_image.shape)
```

```python
plt.figure(figsize=(4,4))
plt.imshow(sample_image)
plt.title("input image")
plt.axis("off")
plt.show()
```

```python
sample_batch= np.expand_dims(sample_image,axis=0)
print("Batch shape:",sample_batch.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Conv2D,ReLU
from tensorflow.keras.models import Model
input_layer= tf.keras.Input(shape=(128,128,3))
conv_layer=Conv2D(filters=32,kernel_size=(3,3),padding="same",activation=None,
                  name="conv_only")(input_layer)
model_before_activation=Model(inputs=input_layer,outputs=conv_layer)

```

```python
import tensorflow as tf
sample_batch = tf.image.resize(sample_batch, (128, 128))
feature_maps_before = model_before_activation.predict(sample_batch)
print("Feature map shape before activation:", feature_maps_before.shape)
```

```python
plt.figure(figsize=(12, 3))

for i in range(4):
    plt.subplot(1, 4, i+1)
    plt.imshow(feature_maps_before[0, :, :, i], cmap='gray')
    plt.title(f"Before Act {i+1}")
    plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
input_layer2=tf.keras.Input(shape=(128,128,3))
conv_output=Conv2D(filters=32,kernel_size=(3,3),padding="same",activation=None,name="conv_raw")(input_layer2)
relu_output=ReLU(name="relu_output")(conv_output)
model_after_activation=Model(inputs=input_layer2,outputs=relu_output)

```

```python
feature_maps_after = model_after_activation.predict(sample_batch)
print("Feature map shape after activation:", feature_maps_after.shape)
```

```python
plt.figure(figsize=(12,3))
for i in range(4):
  plt.subplot(1,4,i+1)
  plt.imshow(feature_maps_after[0,:,:,i],cmap="gray")
  plt.title(f"After ReLU{i+1}")
  plt.axis("off")
  plt.tight_layout()
  plt.show()

```

```python
plt.figure(figsize=(12,6))
for i in range(4):
  plt.subplot(2,4,i+1)
  plt.imshow(feature_maps_before[0:,:,i].squeeze(),cmap="gray")
  plt.title(f"Before ReLU{i+1}")
  plt.axis("off")
  for i in range(4):
    plt.subplot(2,4,i+5)
    plt.imshow(feature_maps_after[0:,:,i].squeeze(),cmap="gray")
    plt.title(f"after ReLU{i+1}")
    plt.axis("off")
    plt.tight_layout()
    plt.show()

```
